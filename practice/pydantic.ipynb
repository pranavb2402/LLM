#### Pydantic python practice
- How to use pydantic BaseModel and Field to define class structure
- It is useful when you want to structurize data models
- It thorws error if type does not match
- It has inbuilt type casting
- It is written in  Rust

```python
from pydantic import BaseModel, Field

class User(BaseModel):
    id: int
    name: str
    email: str=Field(max_length=20, min_length=10)
    password: int

user  = User(id=1, name="John Doe", email="doeexample.com", password=1234)

print(user)
```

#### Typing library practice 
- It works similar to pydantic baseModel but only works while developing, it does not type check in runtime, it simply ignores
- It has some extra modules like union, etc.

```python
from typing import TypedDict

class User(TypedDict):
    id: int
    name: str
    email: str
    password: int

user1 = User(id=1, name="John Doe", email="doeexample.com", password="hh1234")

print(user1)

```

```python
from typing import Union

def add(a: Union[int, float], b: Union[int, float]) -> Union[int, float]:
    return a + b

print(add(1, 2))
print(add(1.0, 2.0))
```

```python
from typing import Optional

def nice_function(a: Optional[int] = 3):
    if a == 3:
        return a
    else:
        return 0

print(nice_function())
```
