#### Tranformer architecture
- Transformer architecure is based on a deep neural network and introduced in 2017 paper 'Attension Is All You Need" 
- It consist of sumbmodules called encoder and decoder.
- Encoder tranforms the input text in vector or numerical representation
- Decoder takes these encoded vectors and generate an output.
- Both encoder and decoder consist of many layers connected by a self attension mechanism
- self-attension mechanism allows model to give importance to different words (token) in terms of sequence related to each other. 
- It helps model to understand the contextual dependecies and long-range relationship to give a relevant output.

### What is Few shot learning and zero-shot learning? 
- In few shot learning, we provide minimal training examples to make model learn.
- In zero-shot learning is capablity of model to generate output even for the completely unseen task.
