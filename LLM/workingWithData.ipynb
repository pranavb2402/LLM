- Deep neural network and also LLMs cannot process raw text data because they are mathematical models. 
- Therefore we need a way to represent word as a continuous valued vectors.
- The process of convertig texts into vectors called embeddings. 

## Let's build Tokenizer

### So, How we can make a tokenizer? (We will be building a tokenizer on small vocab and it is for learning purpose)

So what does tokenizer do? It takes the text string and converts into numerical Ids. 

Follwing are the steps to make our own tokenizer. 
1. Convert training text data into text tokens
2. Map those text tokens to numerical Ids which we can call VOCAB
3. Build Tokenizer class which takes the vocab and initialize encoder and decoder.
4. Encoder tranforms text string into tokenIds and then tokenIds can be converted into string again using decoder.



### Convert text string into text tokens

"Hello, I am here" -> ["Hello", ",", "I", "am", "here"]

We will use a texts store in training.txt file to get all training text data.

```python
# we will be first load text file the-verdict.txt

with open("./assets/the-verdict.txt", "r") as file:
    text = file.read()

print(text)
```

```python
# we will convert that extracted text into text tokens 

tokens = text.split()

print(tokens)

# But we can see texts are only splitted by space. We need to split it by words and punctuations also includes 
# . , ! ? - _ @ "" ' ~ `` etc. and not whitespace

import re

tokens = re.findall(r"[^\W_]+|_|[^\w\s]", text, re.UNICODE)

print(tokens)
```

```python
# Now we need to turn this tokens into numerical tokenIds and map it. we will call it VOCAB 

text_tokens = sorted(set(tokens))

print(text_tokens)

```

### Map text tokens to tokenIDs (It's also called vocab)

["Hi", "i"] -> [123, 456]

```python
# we will create a dictionary to map tokens to tokenIds

vocab = {token: i for i, token in enumerate(text_tokens)}

print(vocab) # Dictionary of text to tokenIds is called vocab
```

### Build Simple Tokenizer class which takes the vocab and initialize encoder and decoder.


```python
class SimpleTokenizer:
    def __init__(self, vocab):
        self.str_to_id = vocab  
        self.id_to_str = {i: s for s, i in vocab.items()}

    def encode(self, text):
        # Convert text string into text tokens
        tokens = re.findall(r"[^\W_]+|_|[^\w\s]", text, re.UNICODE)
        return [self.str_to_id[token] for token in tokens]

    def decode(self, ids):
        return " ".join([self.id_to_str[id] for id in ids])


tokenizer = SimpleTokenizer(vocab)

encoded = tokenizer.encode("I HAD always thought Jack Gisburn rather a cheap genius")

print(encoded)

decoded = tokenizer.decode(encoded)

print(decoded) # we can see encoded tokens and if we pass encoded tokens to decoder it will give us back original text.
```

### Adding special context tokens
In above example we saw if some words are not there in training data vocab then out tokenizer will throw error. How to counter that? one is make a dataset big to acquire all the words (That is not viable all the time) or there is another way to do it.

Add special context tokens like "<'endOfText'>" or "<'unk'>". These tokens we can add in vocab with token Id associated with it. if there is no text in training data while encoding then we will use token "<'unk'>". if there are two different string with different context then we will attach those strings with "<'endOfText'>" like e.g

Hi there I am villain.... You know there is charm in this weather ---> Hi there I am villain.... "<'endOfText'>" You know there is charm in this weather. Because two are different context sentences. Many tokenizer uses this types of special tokens to create encodings and decodings. 

we will extend previously created vocab to include "<'endOfText'>" and "<'unk'>"

```python
#Rebuilding the vocab

preprocessed = sorted(set(tokens))

preprocessed.append("<endOfText>")
preprocessed.append("<unk>")

vocab = {token: i for i, token in enumerate(preprocessed)}

print(vocab) 


class SimpleTokenizer:
    def __init__(self, vocab):
        self.str_to_id = vocab  
        self.id_to_str = {i: s for s, i in vocab.items()}

    def encode(self, text):
        preprocessed = re.split(r'([,.:;?_!"()\']|--|\s)', text)
        preprocessed = [item.strip() for item in preprocessed if item.strip()]
        preprocessed = [item if item in self.str_to_id else "<unk>" for item in preprocessed]
        ids = [self.str_to_id[s] for s in preprocessed]
        return ids

    def decode(self, ids):
        text = " ".join([self.id_to_str[id] for id in ids])
        text = re.sub(r'\s+([,.:;?!"()\'])', r'\1', text)
        return text


tokenizer = SimpleTokenizer(vocab)

encoded = tokenizer.encode("I HAD always thought Jack Gisburn rather a cheap genius")

print(encoded)

decoded = tokenizer.decode(encoded)

print(decoded)
```

```python
encoded = tokenizer.encode("Hello, I am here") # This will throw an error

decoded = tokenizer.decode(encoded)
print(encoded)
print(decoded)

#Hello is not in vocab so it will be replaced with "<unk>"

text1 = "Hello, I am here"
text2=  "I am here"

text2encode = text1 + " <endOfText> " + text2

encoded = tokenizer.encode(text2encode)
decoded = tokenizer.decode(encoded)

print(text2encode)
print(encoded)
print(decoded)

# we can see that "<endOfText>" is added between text1 and text2.


```

### Byte pair encoding
**What BPE is**: It is subword tokenization. Like rather than tokening the word "rare", BPE tokenize "ra" and "re" depending uplon the repeatness of particular subword. subwords are merged based on the frequency of it's in a training dataset. tiktoken has a BPE alogorithm and it is very complex to discuss it here in the scope of this learning. Rather than we will see how we can use it in encoding and decoding.

**Why it matters**: There will be some rare words while encoding some text string, and it is difficult to bring all types of words in vocab. Because while creating vocab it is necessary to make it suitable and light for tokenization.

**Workflow:** During training, count adjacent symbol pairs, merge the most frequent pair, and repeat until the vocabulary reaches the target size. At inference, the same merge rules encode text into token IDs.

You can see following figure to understand what is BPE

!['BPE Image'](./assets/bpe.png) 

```python
import tiktoken

# cl100k_base is used by GPT-4 and GPT-3.5-turbo
encGPT3point5 = tiktoken.get_encoding("cl100k_base")
encGPT2 = tiktoken.get_encoding("gpt2")

print("GPT 3.5 special tokens:", encGPT3point5.special_tokens_set)
print("GPT 2 special tokens:", encGPT2.special_tokens_set)

text = "Hello, world! Byte pair encoding <|endoftext|> handles rare words and 你好."
token_ids = encGPT3point5.encode(text, allowed_special={"<|endoftext|>"})

print("Text:", repr(text))
print("Token IDs:", token_ids)
print("Round-trip:", encGPT3point5.decode(token_ids))
print("Number of tokens:", len(token_ids))
```

As we can see GPT 3.5 uses cl100k_base vocab which has '<|fim_middle|>', '<|fim_suffix|>', '<|endofprompt|>', '<|endoftext|>', '<|fim_prefix|>' these special tokens but GPT 2 uses only one special token <|endoftext|> 

We can see token id of <|endoftext|> is 100257 which is too large and almost coming at the end of all tokens. In vocab spoecial tokens are stored at end of the dictionary.

### Data sampling with a sliding window

As we know tranformers are known to predict the next word or token if provided sequence of texts (tokens). so our next step is to generate a input-target pairs required for training an LLM

In next example we will see how to generate those input-target pairs for the `the-verdict` story

```python
# Read texts from the-verdict.txt

import tiktoken

with open("./assets/the-verdict.txt", "r") as file:
    text = file.read()

# we will use tiktoken tokenizer to encode while story

enc = tiktoken.get_encoding("gpt2")

token_ids = enc.encode(text, allowed_special={"<|endoftext|>"})

# we will store two variables X and Y for input and target

X = []
Y = []

for i in range(len(token_ids) - 1):
    X.append(token_ids[i])
    Y.append(token_ids[i + 1])

print(f"X: {X}")
print(f"Y:    {Y}")

decoded_X = enc.decode(X)
decoded_Y = enc.decode(Y)

print(f"Decoded X: {decoded_X[:100]}")
print(f"Decoded Y:           {decoded_Y[:100]}")

```

```python
# Now we will create a training data of input string and target string
x = []
y = []

context_length = 10

for i in range(1, context_length):
    current_token_ids = token_ids[0:i]
    next_token_id = current_token_ids + [token_ids[i]]
    x.append(current_token_ids)
    y.append(next_token_id)

for i  in range(len(x)):
    print(f"{enc.decode(x[i])} ---------> {enc.decode(y[i])}")
```

We can see how we have generated a input to target string. target strin g always have a next token. Which we can give to train LLM but before that now we have steps to remain to convert tose input-target mapping into embeddings. 

we will use pytorch tensors to create that embeddings. Embeddings helps to create context using vectors (tensors) for LLM.
